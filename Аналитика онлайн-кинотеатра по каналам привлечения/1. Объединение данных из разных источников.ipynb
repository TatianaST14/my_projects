```python
import pandas as pd
```

```python
data_banner = pd.read_excel('data/Копия banner.xlsx')
data_banner['ad_channel'] = 'banner'
data_banner.head()
```

```python
data_native = pd.read_csv('data/Копия native.csv')
data_native = data_native.drop(columns = 'Unnamed: 0')
data_native['ad_channel'] = 'native'
data_native.head()
```

```python
data_organic_traffic = pd.read_csv('data/Копия organic_traffic.csv', header=None)
data_organic_traffic = data_organic_traffic.drop(columns=0)
data_organic_traffic.columns = ['user_id', 'funnel_stage', 'timestamp', 'profit', 'os']
data_organic_traffic['ad_channel'] = 'organic traffic'
data_organic_traffic.head()
```

```python
data_other = pd.read_csv('data/Копия other.csv')
data_other = data_other.drop(columns = 'Unnamed: 0')
data_other['ad_channel'] = 'other'
data_other.head()
```

```python
data_partner_network = pd.read_csv('data/Копия partner_network.csv', header=None, names = ['user_id', 'funnel_stage', 'timestamp', 'profit', 'os'])
data_partner_network['ad_channel'] = 'partner network'
data_partner_network.head()
```

```python
data_rich = pd.read_excel('data/Копия rich.xlsx')
data_rich['ad_channel'] = 'rich'
data_rich.head()
```

```python
data_search = pd.read_csv('data/Копия search.csv')
data_search = data_search.drop(columns = 'Unnamed: 0')
data_search['ad_channel'] = 'search'
data_search.head()
```

```python
data_video = pd.read_csv('data/Копия video.csv')
data_video = data_video.drop(columns = 'Unnamed: 0')
data_video['ad_channel'] = 'video'
data_video.head()
```

```python
df_advertising_information = pd.read_json("data/Копия advertising_information.json")
df_advertising_information.head()
```

```python
df = pd.concat([data_banner, data_native, data_organic_traffic, data_other, \
                data_partner_network, data_rich, data_search, data_video], axis=0)
df.shape
```

```python
all_data = df.merge(df_advertising_information.set_index('ad_channel'), on = 'ad_channel')
all_data
```

```python
all_data = df.merge(df_advertising_information, how = 'left', on = 'ad_channel')
all_data
```

```python
all_data = all_data.sort_values(by = 'timestamp').reset_index(drop=True)
all_data
```

```python
all_data.to_csv('all data.csv', index=False)
```
